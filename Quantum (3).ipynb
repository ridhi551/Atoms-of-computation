```python
#understanding qubits with hadmard gate
from qiskit import QuantumCircuit
q = QuantumCircuit(1,1)
q.h(0)
q.measure(0,0)
q.draw()
```

```python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator()
job = sim.run(q)
result = job.result()
result.get_counts()
```

```python
from qiskit.visualization import plot_histogram
legend = ['Single hadmard gate']

count = {'1': 527, '0': 497}

plot_histogram(count, legend=legend, color='blue', title = "First Histogram")
```

```python
qmix = QuantumCircuit(1,1)
# initial state: 0
# 0 -> 1 -> 0 -> 1
qmix.x(0)
qmix.y(0)
qmix.z(0)
qmix.measure(0,0)

sim=AerSimulator()
job = sim.run(qmix)
result = job.result()
qmix.draw()
result.get_counts()
```

```python
from qiskit.quantum_info import DensityMatrix
import numpy as np
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_qsphere

qc = QuantumCircuit(2)
qc.h([0, 1])
qc.cz(0,1)
qc.ry(np.pi/3, 0)
qc.rx(np.pi/5, 1)
qc.z(1)

matrix = DensityMatrix(qc)
plot_state_qsphere(matrix,
     show_state_phases = True, use_degrees = True)
```
